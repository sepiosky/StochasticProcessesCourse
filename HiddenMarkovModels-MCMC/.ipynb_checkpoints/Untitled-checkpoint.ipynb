```python
def logsumexp(a, axis=None):
    maxel = np.amax(a)
#     if np.isposinf(maxel):
#         return maxel
    return maxel + np.log(np.sum(np.exp(a-maxel), axis=axis)) 

class HMM():
    def __init__(self, n_states, n_obs, implementation="scale"):
        self.n_states = n_states
        self.n_obs = n_obs
        self.implementation = implementation
        # n_x is size of X space
        
    def init_params(self, state_priors=None, trans_prob=None, emit_prob=None):
        self.state_priors = (1/self.n_states)*np.ones(self.n_states) if state_priors is None else state_priors
        self.trans_prob = (1/self.n_states)*np.ones((self.n_states,self.n_states)) if trans_prob is None else trans_prob
        self.emit_prob = (1/self.n_obs)*np.ones((self.n_states,self.n_obs)) if emit_prob is None else emit_prob
        return self
    
    def get_log_params(self):
        return np.log(self.state_priors).copy(), np.log(self.trans_prob).copy(), np.log(self.emit_prob).copy()
        
    def get_params(self):
        return self.state_priors.copy(), self.trans_prob.copy(), self.emit_prob.copy()
    
    def update_params(self, state_priors, trans_prob, emit_prob):
        self.trans_prob = trans_prob
        self.emit_prob = emit_prob
        self.state_priors = state_priors
    
    def forward_eval(self, observation):
        return {'scale':self.forward_eval_scale, 'log':self.forward_eval_log}[self.implementation](observation)
    def backward_eval(self, observation, scales):
        return {'scale':self.backward_eval_scale, 'log':self.backward_eval_log}[self.implementation](observation, scales)
    
    def forward_eval_scale(self, observation):
        seq_length = observation.shape[0]
#         alpha_test = np.zeros((seq_length, self.n_states))
#         scales_test = np.ones(seq_length)
        
#         for i in range(self.n_states):
#             alpha_test[0][i] = self.state_priors[i]*self.emit_prob[i][observation[0]] 
#         scales_test[0] = 1./np.sum(alpha_test[0][:])
#         alpha_test[0][:] *= scales_test[0]
#         for k in range(1,seq_length):
#             for i in range(self.n_states):
#                 alpha_test[k][i] = np.sum([ alpha_test[k-1][j]*self.trans_prob[j][i] for j in range(self.n_states) ])*self.emit_prob[i][observation[k]]
#             scales_test[k] = 1./np.sum(alpha_test[k][:])
#             alpha_test[k][:] *= scales_test[k]
#         log_likelihood = -np.sum(np.log(scales_test)) # in unscaled version np.sum(alpha[-1][:])
        
        alpha = np.zeros((seq_length, self.n_states))
        scales = np.ones(seq_length)

        alpha[0] = self.state_priors * np.array(self.emit_prob)[:,observation[0]]
        scales[0] = 1./np.sum(alpha[0][:])
        alpha[0] *= scales[0]
        for k in range(1,seq_length):
            alpha[k][:] = np.sum(np.array([alpha[k-1]]).T*self.trans_prob, axis=0) * np.array(self.emit_prob)[:,observation[k]]
            #print("alpha[{}] = {}".format(k, alpha[k]))
            scales[k] = 1./np.sum(alpha[k][:])
            #print("scales[{}] to alpha: {}".format(k, scales[k]))
            alpha[k][:] *= scales[k]
            #print("alpha_scaled[{}] = {}".format(k, alpha[k]))
        log_likelihood = -np.sum(np.log(scales)) # in unscaled version np.sum(alpha[-1][:])
        
#         np.testing.assert_almost_equal(alpha, alpha_test)
#         np.testing.assert_almost_equal(scales, scales_test)
                                           
        return log_likelihood, alpha, scales

    def backward_eval_scale(self, observation, scales):
        seq_length = observation.shape[0]
#         beta = np.ones((seq_length, self.n_states))
#         beta[seq_length-1][:] *= scales[seq_length-1]
#         for k in reversed(range(seq_length-1)):
#             for i in range(self.n_states):
#                 beta[k][i] = np.sum([ beta[k+1][j]*self.trans_prob[i][j]*self.emit_prob[j][observation[k+1]] for j in range(self.n_states) ])
#             beta[k][:] *= scales[k]
        
    
        beta = np.ones((seq_length, self.n_states))
        beta[seq_length-1] *= scales[seq_length-1]
        for k in reversed(range(seq_length-1)):
            beta[k] = np.sum(beta[k+1]*self.trans_prob*np.array(self.emit_prob)[:,observation[k+1]], axis=1)
            #print("beta[{}] scale nashode: {}".format(k, beta[k]))
            beta[k] *= scales[k]
            #print("scales[{}] to beta: {}".format(k, scales[k]))
            #print("beta[{}] scale shode: {}".format(k, beta[k]))
        #log_likelihood = -np.sum(np.log(scales)) # in unscaled version np.sum([ self.state_priors[i]*beta[0][i]*self.emit_prob[i][observation[0]] for i in range(self.n_states) ])    
        
        return beta        

    
    def forward_eval_log(self, observation):
        seq_length = observation.shape[0]
        log_trans_prob = np.log(self.trans_prob).copy()
        log_emit_prob = np.log(np.array(self.emit_prob)).copy()
        
        log_alpha = np.zeros((seq_length, self.n_states))
        log_alpha[0] = np.log(self.state_priors) + log_emit_prob[:,observation[0]]
        for k in range(1,seq_length):
            log_alpha[k] = logsumexp(np.array([log_alpha[k-1]]).T + log_trans_prob, axis=1) + log_emit_prob[:,observation[k]]
            #print("berizim roo ham log_alpha[{}] = {}".format(k, log_alpha[k]))
#         log_alpha2 = np.zeros((seq_length, self.n_states))
#         log_alpha2[0] = np.log(self.state_priors) + log_emit_prob[:,observation[0]]
#         for k in range(1,seq_length):
#             for i in range(self.n_states):
#                 log_alpha2[k][i] = logsumexp([ log_alpha2[k-1][j] + np.log(self.trans_prob[j][i]) for j in range(self.n_states) ]) + np.log(self.emit_prob[i][observation[k]])
#                 #print(log_alpha2[k])
        
#         np.testing.assert_almost_equal(log_alpha, log_alpha2)
        
        #print("alpha nan shode ?", np.any( np.isnan(log_alpha) ) )
        return logsumexp(log_alpha[-1]), log_alpha

    def backward_eval_log(self, observation, scales): #scales is useless
        seq_length = observation.shape[0]
        log_trans_prob = np.log(self.trans_prob).copy()
        log_emit_prob = np.log(np.array(self.emit_prob)).copy()
        
        log_beta = np.zeros((seq_length, self.n_states)) #log 1 = 0 (first row is 1 in bckwrd init)
        for k in reversed(range(seq_length-1)):
            log_beta[k] = logsumexp(log_beta[k+1]+log_trans_prob+log_emit_prob[:,observation[k+1]], axis=1)   
                
#         log_beta2 = np.zeros((seq_length, self.n_states))
#         for k in reversed(range(seq_length-1)):
#             for i in range(self.n_states):
#                 log_beta2[k][i] = logsumexp([ log_beta2[k+1][j]+np.log(self.trans_prob[i][j])+np.log(self.emit_prob[j][observation[k+1]]) for j in range(self.n_states) ])
        #log_likelihood = np.exp(logsumexp([ np.log(self.state_priors[i])+log_beta[0][i]+np.log(self.emit_prob[i][observation[0]]) for i in range(self.n_states) ]))
        
#        np.testing.assert_almost_equal(log_beta, log_beta2)
        #print("beta nan shode ?", np.any( np.isnan(log_beta) ) )
        return log_beta
    
    def decode_viterbi(self, observation):
        seq_length = observation.shape[0]
        log_trans_prob, log_emit_prob, log_state_priors = np.ma.log(self.trans_prob).filled(np.NINF) \
                                                        , np.ma.log(self.emit_prob).filled(np.NINF) \
                                                        , np.ma.log(self.state_priors).filled(np.NINF)
                
        delta, par = np.zeros((seq_length, self.n_states)), np.zeros((seq_length, self.n_states), dtype="int64")
        delta[0][:] = [log_state_priors[i]+log_emit_prob[i][observation[0]] for i in range(self.n_states)]
        for k in range(1,seq_length):
            for j in range(self.n_states):
                recursions = [delta[k-1][i]+log_trans_prob[i][j] for i in range(self.n_states)]
                par[k][j] = np.argmax(recursions)
                delta[k][j] = recursions[par[k][j]]+log_emit_prob[j][observation[k]]
        best_state, states_seq = np.argmax(delta[seq_length-1][:]), []
        for k in reversed(range(seq_length)):
            states_seq.append(best_state)
            best_state = par[k][best_state]
        return states_seq[::-1]
    
    
```

```python
class BW():
    def __init__(self, n_states, n_obs, implementation="scale", state_priors=None, trans_prob=None, emit_prob=None):
        self.n_states = n_states
        self.n_obs = n_obs
        self.implementation = implementation
        self.model = HMM(self.n_states, self.n_obs, implementation=self.implementation).init_params(state_priors, trans_prob, emit_prob)
        
    def expectation_log(self, observations, log_alpha, log_beta):
        log_gamma, log_xi = np.zeros((self.n_samples, self.seq_length, self.n_states)), np.zeros((self.n_samples,self.seq_length-1, self.n_states, self.n_states))
        
        _, log_trans_prob, log_emit_prob = self.model.get_log_params()
        print("expectation start")
        for l in range(self.n_samples):
            for k in range(self.seq_length):
                for i in range(self.n_states):
                    log_gamma[l][k][i] = log_alpha[l][k][i]+log_beta[l][k][i]
                    if k < self.seq_length-1 :
                        for j in range(self.n_states):
                            log_xi[l][k][i][j] = log_alpha[l][k][i]+log_trans_prob[i,j]+log_emit_prob[j,observations[l][k+1]]+log_beta[l][k+1][j]
                log_gamma[l][k] -= logsumexp(log_gamma[l][k])
                if k < self.seq_length-1 :
                    log_xi[l][k] -= logsumexp(log_xi[l][k])
        #print("xi nan shode ?", np.any( np.isnan(log_xi) ) )
        #print("gamma nan shode ?", np.any( np.isnan(log_gamma) ) )
        
        print("expectation finish")
        return log_xi, log_gamma
        
    def maximization_log(self, observations, log_xi, log_gamma):
        print("maximiz start")
        new_trans_prob = np.zeros((self.n_states,self.n_states))
        new_emit_prob = np.zeros((self.n_states,self.n_obs))
        new_state_priors = np.zeros(self.n_states)
        
        for i in range(self.n_states):
            new_state_priors[i] = np.exp(logsumexp(log_gamma[:,0,i]))/self.n_samples
            for j in range(self.n_states):
                new_trans_prob[i][j] = np.exp(logsumexp(log_xi[:,:,i,j]) - logsumexp(log_gamma[:,:,i]))
            for j in range(self.n_obs):
                mask = (observations == j).astype(float)
                mask[ mask==0 ] = np.NINF
                mask-=1
                new_emit_prob[i][j] = np.exp(logsumexp(log_gamma[:,:,i]+mask)- logsumexp(log_gamma[:,:,i]))
        new_trans_prob[np.isnan(new_trans_prob)] = 0
        new_emit_prob[np.isnan(new_emit_prob)] = 0
        new_state_priors[np.isnan(new_state_priors)] = 0
        self.model.update_params(new_state_priors.copy(), new_trans_prob.copy(), new_emit_prob.copy())
        del new_trans_prob, new_emit_prob, new_state_priors
        print("maximiz finish")
        
    def calc_likelihood_log(self, observations):
        print("calcll start")
        log_alpha, log_beta = np.zeros((self.n_samples, self.seq_length, self.n_states)), np.zeros((self.n_samples, self.seq_length, self.n_states))
        log_likelihood = 0
        for l in range(self.n_samples):
            log_p, log_alpha[l] = self.model.forward_eval(observations[l])
            #print("log_p: {}".format(log_p))
            log_beta[l] = self.model.backward_eval(observations[l], _)
            log_likelihood += log_p
        print("calcll finish")
        return log_alpha, log_beta, log_likelihood
    
    def fit_log(self, observations):
        log_alpha, log_beta, ll_old = self.calc_likelihood_log(observations)
        while True:
            log_xi, log_gamma = self.expectation_log(observations, log_alpha, log_beta)
            self.maximization_log(observations, log_xi, log_gamma)
            log_alpha, log_beta, ll_new = self.calc_likelihood_log(observations)
            print("ll_old: {}     ll_new: {}".format(ll_old/self.n_samples, ll_new/self.n_samples))
            if abs(ll_new-ll_old) < 0.001 :
                break
            ll_old = ll_new
        return self.model
    

    
    
    
    
    
    
    
    
    def calc_likelihood_scale(self, observations):
        now = datetime.now().strftime("%H:%M:%S")
        print("calc likellihood start: {}",format(now))
        starttime = time.time()
        alphas, betas, scales = np.zeros((self.n_samples, self.seq_length, self.n_states)), np.zeros((self.n_samples, self.seq_length, self.n_states)), np.zeros((self.n_samples, self.seq_length))
        log_likelihood = 0
        for l in range(observations.shape[0]):
            log_p, alphas[l], scales[l] = self.model.forward_eval(observations[l])
            betas[l] = self.model.backward_eval(observations[l], scales[l])
            log_likelihood += log_p
        finishtime = time.time()
        print("calc likellihood finish: {}s".format(finishtime-starttime))   
        return alphas, betas, scales, log_likelihood
    
    def maximization_scale(self, observations, alphas, betas, scales, log_likelihood):
        now = datetime.now().strftime("%H:%M:%S")
        print("Maximization start: {}",format(now))
        starttime = time.time()
                
#         new_trans_prob = np.zeros((self.n_states,self.n_states))
#         new_emit_prob = np.zeros((self.n_states,self.n_obs))        
#         for i in range(self.n_states):
#             for j in range(self.n_states):
#                 trans_normalizer, emit_normalizer = 0, 0
#                 for l in range(self.n_samples):
#                     for k in range(self.seq_length):
#                         if k < self.seq_length-1:
#                             new_trans_prob[i][j] += alphas[l][k][i]*self.trans_prob[i][j]*self.emit_prob[j][observations[l][k+1]]*betas[l][k+1][j]
#                             trans_normalizer += alphas[l][k][i]*betas[l][k][i]/scales[l][k]
#                         if j < self.n_obs:
#                             if observations[l][k]==j:
#                                 new_emit_prob[i][j] += alphas[l][k][i]*betas[l][k][i]/scales[l][k]
#                             emit_normalizer += alphas[l][k][i]*betas[l][k][i]/scales[l][k]
#                 new_trans_prob[i][j] /= trans_normalizer
#                 if j < self.n_obs:
#                     new_emit_prob[i][j] /= emit_normalizer
        
        #self.trans_prob, self.emit_prob = new_trans_prob.copy(), new_emit_prob.copy()
        
        
        state_priors, trans_prob, emit_prob = self.model.get_params()
        new_trans_probb = np.zeros((self.n_states,self.n_states))
        new_emit_probb = np.zeros((self.n_states,self.n_obs))  
        for i in range(self.n_states):
            for j in range(self.n_states):
                trans_num = (alphas[:,0:self.seq_length-1,i] * betas[:,1:self.seq_length,j]) * np.array([np.array(emit_prob[j])[observations[l,1:self.seq_length]] for l in range(self.n_samples)])
                trans_denum = np.multiply(alphas[:,0:self.seq_length-1,i], betas[:,0:self.seq_length-1,i])/scales[:,0:self.seq_length-1]
                new_trans_probb[i][j] = np.sum(trans_num)*trans_prob[i][j]/np.sum(trans_denum)
            
            for j in range(self.n_obs):
                emit_denum = (alphas[:,:,i] * betas[:,:,i])/scales
                emit_num = emit_denum * (observations == j).astype(int)
                new_emit_probb[i][j] = np.sum(emit_num)/np.sum(emit_denum)
        
#         np.testing.assert_almost_equal(self.trans_prob, new_trans_prob)
#         np.testing.assert_almost_equal(self.emit_prob, new_emit_prob)
        
        
        #check_new_priors = [np.sum([np.log(alphas[l][0][i])+np.log(betas[l][0][i]) for l in range(self.n_samples)]) for i in range(self.n_states)]
        new_priors = np.sum(alphas[:,0,:]*betas[:,0,:]/self.n_samples, axis=0)
        #np.testing.assert_almost_equal(new_priors, check_new_priors)
        
        new_priors = new_priors/np.sum(new_priors)
        self.model.update_params(new_priors.copy(), new_trans_probb.copy(), new_emit_probb.copy())
        
        finishtime = time.time()
        print("maximization finish: {}s".format(finishtime-starttime))  
        
    def fit_scale(self, observations):
        alphas, betas, scales, likelihood_old = self.calc_likelihood_scale(observations)
        cnt=0
        while True:
            cnt+=1
            print("----------------------- iter: {} --------------------------".format(cnt))
            self.maximization_scale(observations, alphas, betas, scales, likelihood_old)
            alphas, betas, scales, likelihood_new = self.calc_likelihood_scale(observations)
            #print("assertion alpha sum: {}".format(np.sum(alphas, axis=2)))
            #print("assertion alpha max kol: {}".format( np.amax(alphas)))
            #print("assertion alpha max {}\n alpha min {} ".format(np.amax(alphas, axis=1), np.amin(alphas, axis=1)))
            #print("assertion beta sum: {}".format( np.sum(betas, axis=2)))
            #print("assertion beta min {} \n beta max {} ".format(np.amax(betas, axis=1), np.amin(betas, axis=1)))
#             print("assertion emit :{}".format( np.amax( abs(np.sum(self.emit_prob, axis=1) - 1 ) ) ) )
#             print("assertion trans :{}".format( np.amax( abs(np.sum(self.trans_prob, axis=1) - 1 ) ) ) )
#             print("assertion states :{}".format( abs(np.sum(self.state_priors) - 1 ) ) )
            print("old ll: {}  --  new ll: {}".format(np.exp(likelihood_old/self.n_samples), np.exp(likelihood_new/self.n_samples)))
            print("old log ll: {}  --  new log ll: {}".format(likelihood_old/self.n_samples, likelihood_new/self.n_samples))
            if abs(likelihood_new-likelihood_old) < 0.001 :
                break
            likelihood_old = likelihood_new
        return self.model
    
    
    def fit(self, observations):
        self.n_samples, self.seq_length = observations.shape[0], observations.shape[1]
        return {'scale':self.fit_scale, 'log':self.fit_log}[self.implementation](observations)
        
            
        
```

```python
def steady_state_distribution(transition_matrix):
    n_states = transition_matrix.shape[0]
    A = np.vstack([(transition_matrix-np.diag(np.ones(n_states))).T, np.ones(n_states)])
    B = np.zeros(n_states+1)
    B[-1] += 1
    Q, R = np.linalg.qr(A)
    return np.linalg.solve(R,Q.T@B)
```

```python
def metropolis_hastings(target, proposal, n_samples, burn_in_ratio):
    samples = []
    state = 0
    while len(samples) < n_samples*(1-burn_in_ratio):
        candidate = np.random.choice(np.nonzero(proposal[state])[0])
        q_state_given_given = proposal[candidate][state]/np.sum(proposal[candidate])
        q_candidate_given_state = proposal[state][candidate]/np.sum(proposal[state])
        acceptance_ratio = min(1, (target[candidate]*q_state_given_given)/(target[state]*q_candidate_given_state))
        if np.random.uniform() < acceptance_ratio:
            samples.append(candidate)
            state = candidate
    return samples[int(n_samples*burn_in_ratio):-1]
```
